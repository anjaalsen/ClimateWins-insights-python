### 3.2 Scaling data

```python
# Select columns to scale (excluding 'DATE' and 'MONTH')
columns_to_scale = df.drop(columns=['DATE', 'MONTH']).columns

# Initialize scaler
scaler = StandardScaler()

# Scale only selected columns
df_scaled = df.copy()
df_scaled[columns_to_scale] = scaler.fit_transform(df[columns_to_scale])

# Display the first few rows
df_scaled.head()

```
